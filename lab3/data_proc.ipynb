```python
import numpy as np
import pandas as pd
data = pd.read_csv('data/Steel_industry_data.csv')
display(data)
```

```python
for i in range(1,7):
    column = data.iloc[:,[i]]
    column_name = data.columns[i]
    column = column.to_numpy()
    column = (column - np.min(column)) / (np.max(column) - np.min(column))
    data[column_name] = column
display(data)
```

```python
data['WeekStatus'] = data['WeekStatus'].apply(lambda weekStatus : 0 if 'Weekend' == weekStatus else 1)

def norm_weekday(weekday):
    match weekday:
        case 'Monday': return 0
        case 'Tuesday': return 0.16 
        case 'Wednesday': return 0.33
        case 'Thursday': return 0.5
        case 'Friday': return 0.66
        case 'Saturday': return 0.83
        case 'Sunday': return 1
        case _:
            raise ValueError('Unknown week day')
data['Day_of_week'] = data['Day_of_week'].apply(lambda weekday : norm_weekday(weekday))

def norm_load_type(load_type):
    match load_type:
        case 'Light_Load': return 0
        case 'Medium_Load': return 0.5
        case 'Maximum_Load': return 1
        case _: raise ValueError('Unknown load type')
data['Load_Type'] = data['Load_Type'].apply(lambda load_type : norm_load_type(load_type))

display(data)
data.to_csv('data/data.csv', header=False, index=False,)
```
